```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# 경로 설정
model_path = "./koalpaca_lora_output"
base_model = "beomi/KoAlpaca-Polyglot-5.8B"

# 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(base_model)

# 모델 로드 (메모리 절약 설정)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    device_map="auto",      # 자동 분산
    torch_dtype="auto"      # float16 자동 설정
)

# 프롬프트 구성
prompt = "### 질문:\n주택청약 담보대출이 가능한가요?\n\n### 답변:\n"
inputs = tokenizer(prompt, return_tensors="pt")
inputs.pop("token_type_ids", None)  # ❌ GPT 계열엔 불필요

# 입력 텐서를 모델과 동일한 디바이스로 이동
device = model.device
inputs = {k: v.to(device) for k, v in inputs.items()}

# 텍스트 생성
output = model.generate(
    **inputs,
    max_new_tokens=512,               # 🔹 응답 길이 충분히 확보
    temperature=0.7,                  # 🔹 적당한 창의성
    top_p=0.95,                       # 🔹 확률 누적 상위 토큰만 선택
    do_sample=True,                   # 🔹 샘플링 기반 생성 (greedy 방지)
    repetition_penalty=1.3,          # 🔑 반복 억제 핵심 옵션
    no_repeat_ngram_size=4,          # 🔑 4그램 반복 방지
    pad_token_id=tokenizer.eos_token_id
)


# 출력
generated = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated)
```

```python
from langchain.agents import AgentType, initialize_agent
from langchain.chat_models import ChatOpenAI
from langchain.memory import ConversationBufferMemory
from langchain.tools import StructuredTool
from pydantic import BaseModel
import re
import math

# ✅ 한글 숫자 단위 전처리 (예: 1억5천만원 → 150000000)
def convert_korean_numbers(text: str) -> str:
    units = {"천": 1_000, "만": 10_000, "억": 100_000_000}
    pattern = re.compile(r"(\d+(?:\.\d+)?)([천만억])")

    def replacer(match):
        num = float(match.group(1))
        unit = match.group(2)
        return str(int(num * units[unit]))

    return pattern.sub(replacer, text)

# ✅ 문맥 기반 숫자 단위 보완 (예: 연소득 6000 → 60000000)
def refine_contextual_units(text: str) -> str:
    text = re.sub(r"연\s*소득\s*(\d{3,5})(?!\d|[천만억])", lambda m: str(int(m.group(1)) * 10_000), text)
    return text

# ✅ 최종 전처리 함수
def preprocess(text: str) -> str:
    return convert_korean_numbers(refine_contextual_units(text))

# ✅ 전세대출 한도 계산기
def calculate_jeonse_limit(deposit: int, income: int, credit_grade: str) -> str:
    ratio = 0.8 if credit_grade == "상" else 0.6 if credit_grade == "중" else 0.4
    limit = int(min(deposit * ratio, income * 4))
    return f"당신의 전세대출 한도는 약 {limit:,}원입니다."

# ✅ 월 상환금 계산기
def calculate_monthly_payment(loan_amount: int, annual_interest_rate: float, years: int) -> str:
    r = annual_interest_rate / 100 / 12
    n = years * 12
    monthly = loan_amount * r / (1 - (1 + r) ** -n)
    return f"{loan_amount:,}원을 연 {annual_interest_rate}%의 이자로 {years}년간 갚을 경우, 매월 약 {int(monthly):,}원을 상환하셔야 합니다."

# ✅ 입력 스키마
class JeonseLimitInput(BaseModel):
    deposit: int
    income: int
    credit_grade: str

class MonthlyPaymentInput(BaseModel):
    loan_amount: int
    annual_interest_rate: float
    years: int

# ✅ Tool 정의
tools = [
    StructuredTool.from_function(
        name="jeonse_limit_calc",
        description="전세대출 한도 계산기 (입력: deposit, income, credit_grade)",
        args_schema=JeonseLimitInput,        func=calculate_jeonse_limit
    ),
    StructuredTool.from_function(
        name="monthly_payment_calc",
        description="월 상환금 계산기 (입력: loan_amount, annual_interest_rate, years)",
        args_schema=MonthlyPaymentInput,
        func=calculate_monthly_payment
    )
]

# ✅ LLM + Memory + Agent 설정
llm = ChatOpenAI(model_name="gpt-4", temperature=0)
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

agent = initialize_agent(
    tools=tools,
    llm=llm,
    memory=memory,
    agent=AgentType.OPENAI_FUNCTIONS,
    verbose=True
)

# ✅ 실행 함수
def ask_agent(query: str):
    cleaned = preprocess(query)
    return agent.run(cleaned)

# ✅ 테스트
print(ask_agent("보증금 3억이고 연소득 6천만원, 신용등급은 상인데 전세대출 한도는?"))
print(ask_agent("150000000원을 연 4%로 4년간 갚으면?"))
```

```python
# ✅ 문맥 유지 + 계산 기능 통합된 LangChain 대화형 에이전트

from langchain.chat_models import ChatOpenAI
from langchain.schema.runnable import RunnableLambda, RunnableMap
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.memory import ConversationBufferMemory
from langchain.tools import StructuredTool
from langchain.agents import AgentExecutor, create_openai_functions_agent
from pydantic import BaseModel
import math, re

# ✅ 숫자 단위 전처리 함수
def convert_korean_numbers(text: str) -> str:
    units = {"천": 1_000, "만": 10_000, "억": 100_000_000}

    def repl(match):
        parts = re.findall(r'(\d+)([천만억])?', match.group(0))
        total = 0
        for number, unit in parts:
            total += int(number) * units.get(unit, 1)
        return str(total)

    return re.sub(r'\d+[천만억]?', repl, text)

# ✅ 전세대출 한도 계산 함수
def calculate_jeonse_limit(deposit: int, income: int, credit_grade: str) -> str:
    ratio = 0.8 if credit_grade in ["상", "A"] else 0.6 if credit_grade in ["중", "B"] else 0.4
    limit = int(min(deposit * ratio, income * 4))
    return f"당신의 전세대출 한도는 약 {limit:,}원입니다."

# ✅ 월 상환금 계산 함수
def calculate_monthly_payment(loan_amount: int, annual_interest_rate: float, years: int) -> str:
    r = annual_interest_rate / 100 / 12
    n = years * 12
    monthly = loan_amount * r / (1 - (1 + r) ** -n)
    return f"{loan_amount:,}원을 연 {annual_interest_rate}%의 이자로 {years}년간 갚을 경우, 매월 약 {int(monthly):,}원을 상환하셔야 합니다."

# ✅ Tool 스키마 정의
class JeonseInput(BaseModel):
    deposit: int
    income: int
    credit_grade: str

class MonthlyInput(BaseModel):
    loan_amount: int
    annual_interest_rate: float
    years: int

# ✅ Tool 정의
tools = [
    StructuredTool.from_function(
        name="jeonse_limit_calc",
        description="전세대출 한도 계산기",
        args_schema=JeonseInput,
        func=calculate_jeonse_limit,
    ),
    StructuredTool.from_function(
        name="monthly_payment_calc",
        description="월 상환금 계산기",
        args_schema=MonthlyInput,
        func=calculate_monthly_payment,
    )
]

# ✅ LLM + Memory 구성
llm = ChatOpenAI(model_name="gpt-4", temperature=0)
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

# ✅ 프롬프트 템플릿 구성
prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 금융 상담 전문가입니다. 사용자의 질문을 이해하고 필요한 계산을 도와주세요."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}")
])

# ✅ Agent + Executor 구성
agent = create_openai_functions_agent(llm=llm, tools=tools, prompt=prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, memory=memory, verbose=True)

# ✅ 실행 함수
def ask_agent(query):
    query = convert_korean_numbers(query)
    return agent_executor.invoke({"input": query})["output"]

# ✅ 테스트
print(ask_agent("보증금 3억이고 연소득 6천만원, 신용등급은 상인데 전세대출 한도는?"))
print(ask_agent("그 한도로 4% 이자로 4년 상환하면 매월 얼마야?"))
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.agents import Tool, AgentExecutor, create_openai_functions_agent
from langchain.memory import ConversationBufferMemory
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.tools import StructuredTool
from pydantic import BaseModel
import re
import math

# ✅ 숫자 단위 전처리
def convert_korean_numbers(text: str) -> str:
    units = {"천": 1_000, "만": 10_000, "억": 100_000_000}
    pattern = re.compile(r"(\d+)([천만억])")

    def replace(match):
        num = int(match.group(1))
        unit = match.group(2)
        return str(num * units[unit])

    return pattern.sub(replace, text)

# ✅ 전세대출 한도 계산
def calculate_jeonse_limit(deposit: int, income: int, credit_grade: str) -> str:
    ratio = 0.8 if credit_grade == "상" else 0.6 if credit_grade == "중" else 0.4
    limit = int(min(deposit * ratio, income * 4))
    return f"당신의 전세대출 한도는 약 {limit:,}원입니다."

# ✅ 월 상환금 계산
def calculate_monthly_payment(loan_amount: int, annual_interest_rate: float, years: int) -> str:
    r = annual_interest_rate / 100 / 12
    n = years * 12
    monthly = loan_amount * r / (1 - (1 + r) ** -n)
    return f"{loan_amount:,}원을 연 {annual_interest_rate}%의 이자로 {years}년간 갚을 경우, 매월 약 {int(monthly):,}원을 상환하셔야 합니다."

# ✅ 입력 스키마 정의
class JeonseLimitInput(BaseModel):
    deposit: int
    income: int
    credit_grade: str

class MonthlyPaymentInput(BaseModel):
    loan_amount: int
    annual_interest_rate: float
    years: int

# ✅ Tool 정의
tools = [
    StructuredTool.from_function(
        name="jeonse_limit_calc",
        description="전세대출 한도 계산기 (입력: deposit, income, credit_grade)",
        args_schema=JeonseLimitInput,
        func=calculate_jeonse_limit
    ),
    StructuredTool.from_function(
        name="monthly_payment_calc",
        description="월 상환금 계산기 (입력: loan_amount, annual_interest_rate, years)",
        args_schema=MonthlyPaymentInput,
        func=calculate_monthly_payment
    )
]

# ✅ LLM 및 메모리
llm = ChatOpenAI(model_name="gpt-4", temperature=0)
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

# ✅ 프롬프트 템플릿 (agent_scratchpad 포함)
prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 금융 상담 전문가입니다. 사용자의 질문을 이해하고 필요한 계산을 도와주세요."),
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

# ✅ Agent 및 Executor 구성
agent = create_openai_functions_agent(llm=llm, tools=tools, prompt=prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, memory=memory, verbose=True)

# ✅ 실행 함수
def ask_agent(query: str):
    preprocessed = convert_korean_numbers(query)
    return agent_executor.invoke({"input": preprocessed})["output"]
```
